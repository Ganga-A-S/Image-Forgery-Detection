```python
import keras
import modelCore
import tensorflow as tf
import os
from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt

tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
manTraNet = modelCore.load_trained_model(4, os.getcwd())
```

```python
def read_rgb_image(image_file):
    rgb = cv2.imread(image_file, 1)[..., ::-1]
    return rgb
    
def decode_an_image_array(rgb, manTraNet, dn=1):
    x = np.expand_dims(rgb.astype('float32') / 255. * 2 - 1, axis=0)[:, ::dn, ::dn]
    print(x.shape)
#     return manTraNet.predict(x)[0, ..., 0]

def decode_an_image_file(image_file, manTraNet, dn=1):
    rgb = read_rgb_image(image_file)
    mask = decode_an_image_array(rgb, manTraNet, dn)
    return rgb[::dn, ::dn], mask
```

```python
rgb, mask = decode_an_image_file('/home/neelanjan/Downloads/bigpp.jpg', manTraNet)
# rgb, mask = decode_an_image_file('Neelanjan.jpg', manTraNet)
# plt.figure(figsize=(15, 5))
plt.subplot(141)
plt.title('Original Image')
plt.xticks([])
plt.yticks([])
plt.imshow(rgb)
plt.subplot(142)
plt.title('Predicted Mask')
plt.xticks([])
plt.yticks([])
plt.imshow(mask, cmap='gray')
plt.subplot(143)
plt.title('Overlaid Mask')
plt.xticks([])
plt.yticks([])
plt.imshow(rgb)
plt.imshow(mask, cmap='jet', alpha=.5)
plt.subplot(144)
plt.title('Highlighted Forged Regions')
plt.xticks([])
plt.yticks([])
plt.imshow(np.round(np.expand_dims(mask,axis=-1) * rgb[::1,::1]).astype('uint8'), cmap='jet')

plt.show()
```
